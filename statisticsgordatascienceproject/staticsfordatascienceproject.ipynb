### Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
### Goal_diff_count = GoalsFor - GoalsAgainst

```python
def Goal_diff_count(df):
    Goal_diff_count_df=pd.DataFrame()
    Goal_diff_count_df['Team']=df['Team']
    Goal_diff_count_df['Goal_diff_count']=df['GoalsFor'].astype(int)-df['GoalsAgainst'].astype(int)
    Goal_diff_count_df['Goal_diff_count']=Goal_diff_count_df['Goal_diff_count'].astype(int)
    max_goal_diff=Goal_diff_count_df.nlargest(1,'Goal_diff_count')
    print("The team with maxmium goal difference count is{0} with differnce of{1}"
          .format(max_goal_diff['Team'].to_string(index=False),
                  max_goal_diff['Goal_diff_count'].to_string(index=False)))
    min_goal_diff=Goal_diff_count_df.nsmallest(1,'Goal_diff_count')
    print("The team with minimium goal difference count is{0} with differnce of{1}"
          .format(min_goal_diff['Team'].to_string(index=False),
           min_goal_diff['Goal_diff_count'].to_string(index=False)))
    return Goal_diff_count_df
Goal_return_df=Goal_diff_count(df)
```

### Create a new column with name “Winning Percent” and append it to the data set (5 points)

### Percentage of Winning = (GamesWon / GamesPlayed)*100
### If there are any numerical error, replace it with 0%
### Print the top 5 teams which has the highest Winning percentage

```python
df['Winning Percent']=((df['GamesWon'].astype(int)/df['GamesPlayed'].astype(int))*100)
df['Winning Percent']=df['Winning Percent'].round(decimals=2)
df['Winning Percent']=df['Winning Percent'].fillna('0')
df['Winning Percent']=df['Winning Percent'].astype(float)
df
```

```python
high_prec=df.nlargest(5,'Winning Percent')
print("The top 5 teams which have the highest winning percetages are")
for index,row in high_prec.iterrows():
    print(row['Team'])
```
