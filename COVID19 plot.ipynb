```python
# import libraries
import pandas as pd 
import folium
import numpy as np
import json
```

Dataset link: https://exceljet.net/download-coronavirus-data-to-excel
geojson github repo:https://github.com/johan/world.geo.json

```python
# read the excel file 
df=pd.read_excel(r'C:\Users\prajw\Downloads\covid-19 world eudata.xlsx')
df
```

```python
# check the info of the dataframe 
df.info()
```

```python
# drop the first two rows and unwanted columns
df.drop(df.head(3).index, inplace=True)
df.drop(['Unnamed: 0','Unnamed: 2','Unnamed: 5','Unnamed: 6','Unnamed: 7','Unnamed: 8'],axis=1,inplace=True)
```

```python
df
```

```python
# rename the columns 
df.columns=['country','cases','deaths']
```

```python
df
```

```python
# grouping based on the country
g=df.groupby('country').sum()
g
```

```python
# store the country index because after we reset the index the country index will be deleted 
col=g.index
col
```

```python
# set the new index using np.arange
g=g.set_index(keys=np.arange(g.shape[0]))
g
```

```python
# the stored values are added as a column 
g.insert(loc=0, column='country', value=col)
g
```

```python
# we format the country column 
g['country']=g['country'].str.replace('_',' ')
```

```python
g.info()
```

```python
# cretae a foluim object 
world_map=folium.Map(zoom_state=1,tiles='Mapbox Bright')
# json file path
world_geojson_path=r'C:\Users\prajw\Downloads\world-countries.json'
with open(world_geojson_path, 'r') as j:
     jcont = json.loads(j.read())
```

```python
jcont
```

```python
country_names=[]
# to exctract all the countries name from the geojson file 
for i in range(len(jcont['features'])):
    l=jcont['features'][i]['properties']['name']
    country_names.append(l)
```

```python
# countires that are not in geojson 
d = g[~g['country'].isin(country_names)]
print("Total No countries in the dataset which are not in the geojson are", d.shape[0])
```

```python
# passing paramters to Choropleth 
folium.Choropleth(
          geo_data=world_geojson_path,
          data=g,
          columns=['country','cases'],
          key_on='feature.properties.name',
          fill_color= 'YlOrRd',
          nan_fill_color='white',
          bins=np.linspace(0,g['cases'].max(),num=5),
          legend_name='Covid 19 Confirmed Cases'
).add_to(world_map)
# to show the graph 
world_map
# most of the countries have be in the same range so there is a bit bias in the plotted map
```
